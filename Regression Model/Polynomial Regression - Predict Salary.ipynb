```
# Import package
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# machine learning
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier

# imputing missing values
import missingno as msno

# reading files from local system
from google.colab import files
uploaded = files.upload()
import io

# spliting datasets
from sklearn.model_selection import train_test_split

# getting dummy variables
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```
#Read CSV file
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1]
Y = dataset.iloc[:,-1]
```

```
#Fit data using linear regression model
lin_regressor = linear_model.LinearRegression()
lin_regressor.fit(X,Y)
Y_pred = lin_regressor.predict(X)
```

```
#Transform X to X square
poly_regressor = PolynomialFeatures(4)
X_poly = poly_regressor.fit_transform(X)

#linear model
poly_model = linear_model.LinearRegression()
poly_model.fit(X_poly, Y)
Y_poly_pred = poly_model.predict(X_poly)


#visulize X and Y_pred
fig, ax = plt.subplots()
ax.scatter(X, Y, marker='x', color='red')
#visualize linear regression
ax.plot(X, Y_pred, color = 'blue')
#visualize polynomial regression
ax.plot(X, Y_poly_pred, color = 'black')

#visualize and predict salary using linear regression
ax.plot(6.5, lin_regressor.predict([[6.5]]), 'ro', color = 'blue')

#visualize and predict salary using linear regression
ax.plot(6.5, poly_model.predict(poly_regressor.fit_transform([[6.5]]))
, 'ro', color = 'black')

ax.set_xlabel('Job Level')
ax.set_ylabel('Salary')
```
