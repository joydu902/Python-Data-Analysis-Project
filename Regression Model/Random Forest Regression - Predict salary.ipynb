```
#read csv files
dataset = pd.read_csv('Position_Salaries.csv')
dataset
```

```
#missing values
dataset.isna().sum()
```

```
#split dataset
X = dataset.iloc[:,1].values
Y = dataset.iloc[:,-1].values

X = X.reshape(len(X),1)
Y = Y.reshape(len(Y),1)
```

```
#train random forest regression
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, Y)
```

```
#predict job level at 6.5
regressor.predict([[6.5]])
```

```
#visualize dataset
fig, ax = plt.subplots()
ax.scatter(X, Y, marker='x', color = 'purple')
ax.plot(X, regressor.predict(X))

#higher resolution

X_smooth = np.arange( min(X), max(X), 0.1)
X_smooth = X_smooth.reshape(len(X_smooth), 1)
ax.plot(X_smooth, regressor.predict(X_smooth), color='black')
```
